12. Определим наиболее распространенное мужское и женское имя.

```python
# Фильтруем данные по полу
male_students = df[df['Gender'] == 'Male']
female_students = df[df['Gender'] == 'Female']

# Находим наиболее распространенное имя для каждого пола
print("\nКоличество и процентное соотношение мужских имен:")
display(male_students['Name'].value_counts().reset_index(name='Count'))
display(male_students['Name'].value_counts(normalize=True).mul(100).reset_index(name='Percentage'))


print("\nКоличество и процентное соотношение женских имен:")
display(female_students['Name'].value_counts().reset_index(name='Count'))
display(female_students['Name'].value_counts(normalize=True).mul(100).reset_index(name='Percentage'))
```

14. Преобразуем значения StudyHoursPerWeek в числа с плавающей запятой

15. Вычислим для столбца StudyHoursPerWeek среднее значение, моду, медиану, дисперсию и среднеквадратичное отклонение

```python
# Статистические показатели для StudyHoursPerWeek
study_hours_mean = df['StudyHoursPerWeek'].mean()
study_hours_mode = df['StudyHoursPerWeek'].mode()
study_hours_median = df['StudyHoursPerWeek'].median()
study_hours_variance = df['StudyHoursPerWeek'].var()
study_hours_std = df['StudyHoursPerWeek'].std()

print(f"\nСреднее значение StudyHoursPerWeek: {study_hours_mean:.4f}")
print(f"Мода StudyHoursPerWeek:")
display(study_hours_mode)
print(f"Медиана StudyHoursPerWeek: {study_hours_median:.4f}")
print(f"Дисперсия StudyHoursPerWeek: {study_hours_variance:.4f}")
print(f"Среднеквадратичное отклонение StudyHoursPerWeek: {study_hours_std:.4f}")
```
