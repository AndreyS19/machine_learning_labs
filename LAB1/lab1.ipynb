###Датасет - student_performance.csv
###Этот набор данных предоставляет информацию о результатах студентов и предназначен для оценки и прогнозирования результатов учащихся на основе различных факторов, влияющих на академическую успеваемость.
###Описание полей:
*   StudentID – уникальный идентификатор для каждого
студента
*   Name – имя студента.
*   Gender – пол студента (мужчина/женщина).
*   AttendanceRate – процент занятий, посещенных студентом

*   ParentalSupport: качественная оценка уровня поддержки,оказываемой родителями студента (высокий/средний/низкий).
*   FinalGrade: итоговая оценка студента (из 100), которая служит целевой переменной для прогнозирования




**1. Импортируйм необходимые библиотеки.**

```python
import pandas as pd
```

**2. Импортируем набор данных из файла student_performance.csv и помещаем его в датафрейм.**

```python
df = pd.read_csv('drive/MyDrive/Colab Notebooks/student_performance.csv')
```

**3. Выводим 16 первых строк датафрейма.**

```python
print("Первые 16 строк набора данных:")
display(df.head(16))
```

**4. Узнаем, сколько строк и столбцов в наборе данных.**

```python
print("\nКоличество строк и столбцов в наборе данных:")
display(df.shape)
```

**5. Получим сводную информацию по датафрейму.**

```python
print("\nСводная информация по датафрейму:")
display(df.info())
```

**6. Определим, что является индексом этого набора данных и проиндексируем его по полю StudentID.**

```python
print("\nИндекс до изменения:")
display(df.index)

df.set_index('StudentID', inplace=True)

print("\nИндекс после изменения:")
display(df.index)
```

**7. Определим, какие признаки в наибольшей степени предсказывают успеваемость студентов.**

```python
correlation_matrix = df.corr(numeric_only=True)

# Корреляция с 'FinalGrade'
final_grade_correlations = correlation_matrix['FinalGrade'].drop('FinalGrade')

print("\nПризнаки, наиболее сильно предсказывающие успеваемость (по корреляции с FinalGrade):")
display(final_grade_correlations.sort_values(ascending=False))
```

**8. Выявим степень, в которой регулярная посещаемость влияет на итоговые оценки.**


```python
correlation_attendance_finalgrade = df['AttendanceRate'].corr(df['FinalGrade'])

print(f"\nКоэффициент корреляции между AttendanceRate и FinalGrade: {correlation_attendance_finalgrade:.2f}")
```

**9. Оценим, оказывает ли участие во внеклассных мероприятиях положительное или отрицательное влияние на успеваемость.**

```python
correlation_extracurricular_finalgrade = df['ExtracurricularActivities'].corr(df['FinalGrade'])

print(f"\nКоэффициент корреляции между ExtracurricularActivities и FinalGrade: {correlation_extracurricular_finalgrade:.2f}")
```

**10. Как различные уровни родительской поддержки влияют на успеваемость студентов?**

```python
average_grade_by_parental_support = df.groupby('ParentalSupport')['FinalGrade'].mean()

print("\nСредняя итоговая оценка по уровням родительской поддержки:")
display(average_grade_by_parental_support)
```

**11. Какой вклад в успеваемость вносит время, потраченное на учебу?**

```python
correlation_study_hours_finalgrade = df['StudyHoursPerWeek'].corr(df['FinalGrade'])

print(f"\nКоэффициент корреляции между StudyHoursPerWeek и FinalGrade: {correlation_study_hours_finalgrade:.2f}")
```

**12. Определим наиболее распространенное мужское и женское имя.**

```python
# Фильтруем данные по полу
male_students = df[df['Gender'] == 'Male']
female_students = df[df['Gender'] == 'Female']

# Находим наиболее распространенное имя для каждого пола
print("\nКоличество и процентное соотношение мужских имен:")
display(male_students['Name'].value_counts().reset_index(name='Count'))
display(male_students['Name'].value_counts(normalize=True).mul(100).reset_index(name='Percentage'))


print("\nКоличество и процентное соотношение женских имен:")
display(female_students['Name'].value_counts().reset_index(name='Count'))
display(female_students['Name'].value_counts(normalize=True).mul(100).reset_index(name='Percentage'))
```

**13. Проиндексируем датафрейм по полю gender и определим раздельно
наибольшее и наименьшее значение процента посещенных занятий.**

```python
print("\nИндекс до изменения (для задачи 13):")
display(df.index)

df_gender_indexed = df.set_index('Gender')

print("\nИндекс после изменения (для задачи 13):")
display(df_gender_indexed.index)

# Находим наибольшее и наименьшее значение AttendanceRate для каждого пола
attendance_by_gender = df_gender_indexed.groupby('Gender')['AttendanceRate']

print("\nНаибольшее значение AttendanceRate по полу:")
display(attendance_by_gender.max())

print("\nНаименьшее значение AttendanceRate по полу:")
display(attendance_by_gender.min())

# Вернем исходный индекс для последующих задач
df.reset_index(inplace=True)
df.set_index('StudentID', inplace=True)
```

**14. Преобразуем значения StudyHoursPerWeek в числа с плавающей запятой**

```python
df['StudyHoursPerWeek'] = df['StudyHoursPerWeek'].astype(float)

print("\nТипы данных после преобразования StudyHoursPerWeek:")
display(df.info())
```

**15. Вычислим для столбца StudyHoursPerWeek среднее значение, моду, медиану, дисперсию и среднеквадратичное отклонение**

```python
# Статистические показатели для StudyHoursPerWeek
study_hours_mean = df['StudyHoursPerWeek'].mean()
study_hours_mode = df['StudyHoursPerWeek'].mode()
study_hours_median = df['StudyHoursPerWeek'].median()
study_hours_variance = df['StudyHoursPerWeek'].var()
study_hours_std = df['StudyHoursPerWeek'].std()

print(f"\nСреднее значение StudyHoursPerWeek: {study_hours_mean:.4f}")
print(f"Мода StudyHoursPerWeek:")
display(study_hours_mode)
print(f"Медиана StudyHoursPerWeek: {study_hours_median:.4f}")
print(f"Дисперсия StudyHoursPerWeek: {study_hours_variance:.4f}")
print(f"Среднеквадратичное отклонение StudyHoursPerWeek: {study_hours_std:.4f}")
```
